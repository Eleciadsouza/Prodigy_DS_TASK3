```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sn
import os
```

```python
cwd=os.getcwd()
df=pd.read_csv(cwd+"\\bank-full.csv",sep=';')
df.head()
```

```python
data=df['age;"job";"marital";"education";"default";"balance";"housing";"loan";"contact";"day";"month";"duration";"campaign";"pdays";"previous";"poutcome";"y"'].str.split(';',expand=True)
data
```

```python
a=df.columns.str.split(';')
a
```

```python
b=['age', 'job', 'marital', 'education', 'default', 'balance', 'housing', 'loan', 'contact', 'day', 'month', 'duration', 'campaign', 'pdays', 'previous', 'poutcome', 'y']
```

```python
data.columns=b[:len(data.columns)]
data.head()
```

```python
data.info()
```

```python
print(data.isna().sum())
```

```python
d1=['age','duration','campaign','pdays','previous','day']
d2=['balance']
```

```python
data[d1]=data[d1].astype(int)
data[d2]=data[d2].astype(float)
```

```python
data.describe()
```

```python
data.columns
```

```python
#categotrical to numerical values
cn=['age','duration','campaign','pdays','previous','day','balance','y']
cat_data=[col for col in data.columns if col not in cn] 
data=pd.get_dummies(data,columns=cat_data)
data
```

```python
#train test split
from sklearn.model_selection import train_test_split

X=data.drop('y',axis=1)
y=data['y']
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(ccp_alpha=0.001)
model.fit(x_train,y_train)
```

```python
#make predictions
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
accuracy=accuracy_score(y_test,y_pred)
accuracy
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
from sklearn.tree import plot_tree 
plt.figure(figsize=(20,10))
plot_tree(model, filled=True, feature_names=X.columns, class_names=['No', 'Yes'])
plt.show()
```
